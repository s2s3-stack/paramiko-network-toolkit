编写一个脚本，将函数的文档保存到文本文件中。
编写一个Python程序，一次性打印多个函数的文档。

```python
doc_list = [] #创建一个空列表用来存储用户输入的模块文档字符串
filename = 'file.txt'
import os
while True: #循环输入模块文档字符串，直到用户输入'quit'为止
    doc = input("Input your document: ")
    document = doc_list.append(doc)
    if doc == 'quit':
        break
    if not os.path.exists(filename): #检查文件是否存在，如果不存在则创建新文件并写入内容
        for doc in document:
            with open(filename,'w',encoding="utf-8") as f:
                f.write(f"写入模块{doc.__doc__}的内容\n")
        print(f"文件{filename}创建成功")
    else:
        with open(filename,'a',encoding="utf-8") as f:
            f.write(f"写入模块{doc.__doc__}的内容\n")
        print(f"文件{filename}追加成功")
        
    
print(doc_list)
```

计算两个日期之间的小时差值

```python
from datetime import datetime

def calculate_hours_between_dates():
    # 获取用户输入的日期时间（格式：YYYY-MM-DD HH:MM）
    dt1_str = input("请输入第一个日期时间（格式：YYYY-MM-DD）：")
    dt2_str = input("请输入第二个日期时间（格式：YYYY-MM-DD）：")
    
    # 转换为datetime对象
    dt1 = datetime.strptime(dt1_str, "%Y-%m-%d") #计算日期时间差值
    dt2 = datetime.strptime(dt2_str, "%Y-%m-%d")
    
    # 计算小时差
    hours_diff = abs((dt2 - dt1) * 24) #转换为小时
    
    print(f"两个日期时间之间相差 {hours_diff:.2f} 小时") #保留两位小数

# 调用函数  
calculate_hours_between_dates()
```

编写一个脚本，检查给定日期是否为周末。

```python
from datetime import datetime
days = input("请输入一个日期（格式：YYYY-MM-DD）：")

date_obj = datetime.strptime(days, "%Y-%m-%d") #转换为datetime对象
weekday = date_obj.weekday() #获取星期几，0-周一，6-周日
if weekday == 6 or weekday == 5: #周六或周日
    print(f"{days}是周末")
else:
    print(f"{days}是工作日")
```

编写一个程序，接受一个数字并打印该数字之前的所有偶数或奇数。

```python
# 接收用户输入的数字
num = int(input("请输入一个整数: "))
# 询问要打印偶数还是奇数
choice = input("请选择打印偶数还是奇数（输入'偶数'或'奇数'）: ")

print(f"数字 {num} 之前的所有{choice}是:")
# 遍历num之前的数（从1到num-1）
for i in range(1, num):
    # 用位运算判断奇偶（也可以用i % 2，这里沿用之前的位运算方式）
    if choice == "偶数" and not (i & 1): #not 指定取反
        print(i, end=" ")
    elif choice == "奇数" and (i & 1):
        print(i, end=" ") #end 参数用于控制打印时的结尾符号，默认是换行符，这里改为一个空格，使输出在一行显示
```

编写一个 Python 程序，获取给定字符串前 2 个字符的 n 份（n 为非负整数）。如果字符串长度小于 2，则返回整个字符串的 n 份。
语法:序列[起始索引:结束索引:步长]

```python
def substring_copy(str,n):
    flen = 2
    if flen > len(str): # 检查flen是否大于字符串长度
        flen = len(str) #调整flen为字符串长度，避免索引越界
    substr = str[:flen] #调整获取前flen个字符(取反设置::-1)
    result = ""
    for i in range(n):
        result += substr #+=代表连接字符串
    return result
print(substring_copy("Hello", 3))
print(substring_copy("ab", 4))
```

编写一个函数，该函数接受一个字符串和一个整数n，并返回由大写和小写字母交替组成的、重复n次的模式。

```python
def alternate_case_repeat(s, n):
    # 步骤1：生成大小写交替的基础模式
    alternate_pattern = ""
    for idx, char in enumerate(s):
        """
        编写一个函数，该函数接受一个字符串和一个整数n，并返回由大写
        和小写字母交替组成的、重复n次的模式。
        """
        if idx % 2 == 0:  # 索引为偶数（第1、3、5...个字符）→ 大写
            alternate_pattern += char.upper()
        else:  # 索引为奇数（第2、4、6...个字符）→ 小写
            alternate_pattern += char.lower()
    
    # 步骤2：将基础模式重复n次
    return alternate_pattern * n

# 测试示例
print(alternate_case_repeat("hello",  2))  
# 基础模式：H e L l O → "HeLlO"；重复2次 → "HeLlOHeLlO"

print(alternate_case_repeat("abc", 3))    
# 基础模式：A b C → "AbC"；重复3次 → "AbCAbCAbC"
```

编写一个函数，查找列表中指定值的第一个出现位置，并返回其索引

```python
def ser_first_index(str,index):
    for ind,value in enumerate(str):
        if value == index:
            return ind
    return -1  # 如果未找到，返回-1表示不存在
print(ser_first_index("hello", "l"))  # 输出: 2
print(ser_first_index("world", "z"))
```

编写一个脚本，使用星号(*)显示直方图，但将其右对齐。

```python
def right_aligned_histogram(num_list):
    # 步骤1：获取列表最大值，作为右对齐的基准宽度
    max_num = max(num_list) if num_list else 0
    
    # 步骤2：遍历列表，生成右对齐的星号串
    for num in num_list:
        # 生成num个星号
        star_str = "*" * num
        # rjust(max_num)：在max_num宽度内右对齐（左侧补空格）
        right_aligned_str = star_str.rjust(max_num) 
        print(right_aligned_str)

# 测试示例
test_list = [2, 5, 3, 8, 4]
right_aligned_histogram(test_list)
```
