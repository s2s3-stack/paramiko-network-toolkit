加载表格，定位sheet表，向表格输出内容

```python
import openpyxl
wb = openpyxl.load_workbook(r"Ceshi1.xlsx") #加载一个已有的工作簿
# print(wb.sheetnames) #获取工作簿中所有工作表的名称，返回一个列表
sheet = wb['Sheet1'] #通过工作表名称获取工作表对象
sheet['A1'] = "Hello, World!" #向单元格A1写入数据
wb.save("Ceshi1.xlsx") #保存工作簿到指定的文件
sheet['A1'].value #读取单元格A1的数据
sheet.cell(row=1, column=2) #通过行号和列号获取单元格对象
sheet.cell(row=1, column=2).value #读取单元格B1的数据
sheet.max_row #获取工作表的最大行数
sheet.max_column #获取工作表的最大列数
```

获取行和列

```python
import openpyxl
wb = openpyxl.load_workbook('Ceshi1.xlsx')
sheet = wb['Sheet1']
sheet['A1':'C3'] #获取A1到C3范围内的所有单元格对象，返回一个二维元组
# for row in sheet['A1':'C3']: #遍历二维元组中的每一行
#     for cell in row: #遍历每一行中的每个单元格对象
#         print(cell.coordinate, cell.value) #打印单元格的坐标和值
#     print("----")
for cell_obj in list(sheet.columns)[1]: #遍历工作表的第二列（B列）中的每个单元格对象
    print(cell_obj.coordinate, cell_obj.value) #打印单元格的坐标和值
```

修改表格标题名称

```python
import openpyxl
wb = openpyxl.load_workbook('Ceshi1.xlsx')
sheet = wb.active
sheet.title = "MySheet" #修改工作表的名称
wb.save('Ceshi1.xlsx') #保存工作簿到指定的文件
wb.sheetnames #获取工作簿中所有工作表的名称，返回一个列表
```

```python
import openpyxl
wb = openpyxl.load_workbook('Ceshi1.xlsx',data_only=True) #加载一个已有的工作簿，并启用data_only参数
sheet = wb['Sheet']
sheet['E1'] = '=SUM(A1:D1)' #在单元格E1中写入一个求和公式
wb.save('Ceshi1.xlsx') #保存工作簿到指定的文件
result = sheet['E1'].value #读取单元格E1中的公式字符串
print(result)
```
